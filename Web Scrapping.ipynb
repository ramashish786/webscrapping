```python
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://www.espncricinfo.com/records/trophy/batting-list-hundreds/indian-premier-league-117"
```

```python
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import re
```

```python
columns = []
rows = []
if response.status_code == 200:
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    table = soup.find('table')
    
    if table:
        
        for row in table.find_all('tr'):
           
            cells = row.find_all('th')  
            for cell in cells:
                columns.append(cell.get_text())
            
        for row in table.find_all('tr'):
            
            cells = row.find_all("td")
            for cell in cells:
                 rows.append(cell.text)   
    else:
        print('No table found on the webpage.')

else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```

```python
columns = rows[0:12]
```

```python
numArray = np.array(rows[12:])
```

```python
frame = []
```

```python
for i in range(len(columns),len(numArray),len(columns)):
    frame.append(numArray[i-len(columns):i])
```

```python
raw_data = pd.DataFrame(frame,columns=columns)
```
